## Concept: Multimodal LLMs

**Vision-language models** take images and text and output text (e.g. captioning, VQA). **CLIP** aligns image and text embeddings. **BLIP**, **LLaVA**, **Flamingo**-style models combine vision encoder + LLM. We use a small VLM or CLIP for illustration.
