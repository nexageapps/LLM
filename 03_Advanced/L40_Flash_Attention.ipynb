## Concept: Flash Attention

**Flash Attention** (Dao et al.) computes attention in blocks (tiles), keeps only O(N) memory instead of O(N^2) for the full attention matrix, and uses SRAM efficiently. Supported in PyTorch 2.0+ (scaled_dot_product_attention with backend flash) and via flash-attn package. Requires compatible GPU (e.g. A100, Ampere+).
