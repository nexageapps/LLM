## Concept: Alignment

**Alignment** = train or steer models toward human values: **helpful** (follow instructions), **harmless** (refuse harmful requests), **honest** (avoid deception). Methods: RLHF (L25), Constitutional AI, red-teaming, safety classifiers, and prompt-level controls.

## Part 1: Safety/Content Classification

---

```python
# Example: use a content moderation or safety model to score outputs
classifier = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-offensive", top_k=1)
texts = ["That is a great idea!", "I will hurt you."]
for t in texts:
    out = classifier(t)
    print(f"{t[:40]}... -> {out}")
```
