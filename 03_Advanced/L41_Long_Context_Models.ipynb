## Concept: Long Context

**Limits**: Standard attention is O(L^2). **Solutions**: (1) **RoPE** (rotary), **ALiBi** (relative bias) allow extrapolation; (2) **Sparse/local** attention reduces cost; (3) **Flash Attention** reduces memory. Many models (Llama, Mistral) support 4K–32K+ with proper config.

## Part 1: Model Max Position Embeddings

---

```python
from transformers import GPT2Config
config = GPT2Config.from_pretrained("gpt2")
print(f"GPT-2 default max position: {config.n_positions}")
print("Long-context models (e.g. Llama 2, Mistral) support 4K-32K+ with RoPE.")
```

## Part 2: Handling Long Sequences (Truncation and Sliding)

---

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
long_text = "Hello. " * 500
tokens = tokenizer(long_text, truncation=True, max_length=1024, return_tensors="pt")
print(f"Truncated to {tokens['input_ids'].shape[1]} tokens.")
print("For full long context: use a model with larger max_position_embeddings and sufficient memory.")
```

## Part 3: RoPE / ALiBi (Concept)

---

```python
print("RoPE: rotary position embeddings in attention; supports length extrapolation.")
print("ALiBi: attention with additive bias by distance; no position embedding.")
print("Both allow training at one length and inference at longer (with possible quality drop).")
```

## Key Takeaways

1. Standard models have fixed max length; long-context models use RoPE/ALiBi and optimizations.
2. Truncation or chunking is the fallback when model length is limited.
3. Flash Attention and sparse attention enable practical long-context training.
