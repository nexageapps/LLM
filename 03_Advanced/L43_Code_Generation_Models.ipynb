## Concept: Code LMs

**Codex** (OpenAI), **CodeGen**, **StarCoder**, **InCoder** are trained on code and natural language. Use for: completion, docstring-to-code, code-to-code (translate/fix), or text-to-code. Same interface as causal LMs with code-specific prompts and tokenizers.

```python
!pip install transformers torch -q
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import warnings
warnings.filterwarnings('ignore')
print("Libraries loaded.")
```

## Part 1: Code Completion with a Code Model

---

```python
# Use a small code model (larger: Salesforce/codegen-350M-mono, bigcode/starcoderbase)
model_name = "Salesforce/codegen-350M-mono"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

prompt = "def factorial(n):"
inputs = tokenizer(prompt, return_tensors="pt")
out = model.generate(**inputs, max_new_tokens=80, do_sample=False, pad_token_id=tokenizer.eos_token_id)
print(tokenizer.decode(out[0], skip_special_tokens=True))
```

## Part 2: Docstring-to-Code Prompt

---

```python
prompt2 = "# Write a function that returns the sum of two numbers\ndef add(a, b):"
inputs2 = tokenizer(prompt2, return_tensors="pt")
out2 = model.generate(**inputs2, max_new_tokens=40, do_sample=False, pad_token_id=tokenizer.eos_token_id)
print(tokenizer.decode(out2[0], skip_special_tokens=True))
```
