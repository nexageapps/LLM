## Concept: Model Parallelism

When a single model doesn't fit on one GPU: **Pipeline parallelism** splits layers across GPUs (e.g. layers 0–7 on GPU0, 8–15 on GPU1). **Tensor parallelism** splits a single layer (e.g. attention heads) across GPUs. **device_map** in HuggingFace places modules on devices; accelerate supports "auto".

## Part 2: Pipeline Parallelism (Concept)

---

```python
# Pipeline: forward pass in stages. Micro-batches flow through stages.
# Example: Stage 0 (layers 0-3) on GPU0, Stage 1 (layers 4-7) on GPU1.
# Libraries: PyTorch PiPPy, DeepSpeed pipeline, or manual nn.Module split.
print("Pipeline parallelism: sequential stages on different GPUs; reduces per-GPU memory.")
```

## Part 3: Tensor Parallelism (Concept)

---

```python
# Tensor parallel: split matrix (e.g. QKV) across devices; all-gather for output.
# Used in Megatron-LM, DeepSpeed. Requires custom layer implementations.
print("Tensor parallelism: split single layer across GPUs; more communication, finer grain.")
```

## Key Takeaways

1. device_map="auto" (with accelerate) spreads layers across available devices.
2. Pipeline parallelism = layer-wise split; tensor = within-layer split.
3. Combine with data parallelism (e.g. DeepSpeed 3D) for very large training.
