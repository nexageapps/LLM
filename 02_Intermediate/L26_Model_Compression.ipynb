## Concept: Compression

**Quantization**: store weights in lower precision (FP16, INT8, INT4) to reduce memory and often speed. **Pruning**: remove low-importance weights. **Distillation**: train a small student to mimic a large teacher. We focus on quantization here.

---

```python
!pip install transformers torch accelerate bitsandbytes -q
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import warnings
warnings.filterwarnings('ignore')
print("Libraries loaded.")
```

## Part 1: Load in FP16 (Half Precision)

---

```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model_fp16 = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16)
n_params = sum(p.numel() for p in model_fp16.parameters())
print(f"Params: {n_params/1e6:.2f}M. FP16 halves memory vs FP32.")
```

## Part 2: INT8 Quantization with bitsandbytes

Load model with 8-bit quantization for smaller memory footprint.

---

```python
model_8bit = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_8bit=True,
    device_map="auto",
)
print("Model loaded in 8-bit. Useful for large models on limited GPU memory.")
```

## Part 3: Compare Size (Conceptual)

---

```python
# FP32: 4 bytes/param, FP16: 2, INT8: 1
print("Approximate size: FP32 4B/param, FP16 2B/param, INT8 1B/param")
print("For 1B params: FP32 ~4GB, FP16 ~2GB, INT8 ~1GB")
```
