## Concept: BERT Family

**BERT** (encoder-only) is pre-trained with MLM and NSP. **RoBERTa** drops NSP, uses more data. **DistilBERT** is smaller/faster via distillation. **ALBERT** shares parameters. Use them for classification, NER, QA, and sentence embeddings.

---

```python
!pip install transformers torch -q
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification, AutoModelForTokenClassification, AutoModelForQuestionAnswering
import warnings
warnings.filterwarnings('ignore')
print("Libraries loaded.")
```

## Part 1: Classification with BERT and DistilBERT

---

```python
classifier_bert = pipeline("sentiment-analysis", model="bert-base-uncased")
classifier_distil = pipeline("sentiment-analysis", model="distilbert-base-uncased")
texts = ["This movie is great!", "I did not like it."]
print("BERT:", classifier_bert(texts))
print("DistilBERT:", classifier_distil(texts))
```

## Part 2: NER with BERT-style Model

---

```python
ner = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")
print(ner("Alice works at Google in Paris."))
```

## Part 3: Question Answering

---

```python
qa = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "The Eiffel Tower is in Paris. It was built in 1889."
print(qa(question="Where is the Eiffel Tower?", context=context))
print(qa(question="When was it built?", context=context))
```

## Part 4: Compare Model Sizes

---

```python
for name in ["bert-base-uncased", "roberta-base", "distilbert-base-uncased"]:
    model = AutoModelForSequenceClassification.from_pretrained(name, num_labels=2)
    n = sum(p.numel() for p in model.parameters())
    print(f"{name}: {n/1e6:.1f}M params")
```
