## Concept: T5

**T5** frames tasks as "text in → text out" with task prefixes (e.g. "translate English to German:", "summarize:"). Encoder-decoder; good for translation, summarization, QA, and other conditional generation.

---

```python
!pip install transformers torch -q
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import warnings
warnings.filterwarnings('ignore')
print("Libraries loaded.")
```

## Part 1: Summarization with T5

---

```python
summarizer = pipeline("summarization", model="t5-small")
text = "The Eiffel Tower is a landmark in Paris. It was built in 1889. Millions of people visit every year. It is one of the most famous structures in the world."
print(summarizer(text, max_length=30, min_length=10))
```

## Part 2: Translation (Text-to-Text)

---

```python
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
tokenizer = AutoTokenizer.from_pretrained("t5-small")
input_text = "translate English to German: The weather is nice today."
inputs = tokenizer(input_text, return_tensors="pt")
outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

## Part 3: Question Answering (T5 style)

---

```python
input_qa = "question: What is the capital of France? context: Paris is the capital of France."
inputs = tokenizer(input_qa, return_tensors="pt")
out = model.generate(**inputs, max_new_tokens=10)
print("Answer:", tokenizer.decode(out[0], skip_special_tokens=True))
```
