## Concept: Model Merging

**Weight averaging**: same-architecture models can be averaged (e.g. (W1+W2)/2). **TIES**: trim, then merge LoRA with sign disagreement resolution. **DARE**: drop random adapter weights, then scale and merge. Improves multi-task or robustness. We do simple averaging.

---

```python
!pip install transformers torch -q
import torch
from transformers import AutoModelForCausalLM
import copy
import warnings
warnings.filterwarnings('ignore')
print("Libraries loaded.")
```

## Part 1: Load Two Checkpoints (Same Arch)

---

```python
model_name = "gpt2"
model1 = AutoModelForCausalLM.from_pretrained(model_name)
model2 = AutoModelForCausalLM.from_pretrained(model_name)
state1 = model1.state_dict()
state2 = model2.state_dict()
print("Two identical base models loaded (for demo). In practice use two fine-tuned checkpoints.")
```

## Part 2: Average Weights

---

```python
merged = copy.deepcopy(model1)
merged_state = merged.state_dict()
for k in merged_state:
    if k in state2:
        merged_state[k] = (state1[k].float() + state2[k].float()) / 2
merged.load_state_dict(merged_state)
print("Merged model = average of model1 and model2 weights.")
```

## Part 3: TIES / DARE (Concept)

---

```python
print("TIES: trim small delta weights, resolve sign conflicts, then merge.")
print("DARE: randomly zero out adapter params, scale, then merge. Use library 'mergekit' or similar.")
```
