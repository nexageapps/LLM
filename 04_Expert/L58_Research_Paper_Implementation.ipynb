## Concept: Paper Implementation

**Steps**: (1) Read abstract and method; (2) List equations and hyperparameters; (3) Implement in code (start small); (4) Run on same or similar setup; (5) Document differences. We implement a simple idea from the literature: **GELU** activation (used in BERT/transformers) and compare to ReLU.

---

```python
import torch
import torch.nn as nn
import math
import warnings
warnings.filterwarnings('ignore')
print("Libraries loaded.")
```

## Part 1: Implement GELU from Paper

---

```python
# GELU: x * Phi(x) where Phi is standard normal CDF. Paper: "Gaussian Error Linear Units" (Hendrycks & Gimpel).
def gelu_paper(x):
    return 0.5 * x * (1 + torch.tanh(math.sqrt(2/math.pi) * (x + 0.044715 * x**3)))

x = torch.randn(4, 8)
print("GELU (paper approx) output shape:", gelu_paper(x).shape)
print("PyTorch GELU:", torch.nn.functional.gelu(x).shape)
```

## Part 2: Compare to ReLU in a Tiny Net

---

```python
class TinyMLP(nn.Module):
    def __init__(self, use_gelu=True):
        super().__init__()
        self.fc = nn.Linear(32, 16)
        self.act = gelu_paper if use_gelu else nn.ReLU()
        self.fc2 = nn.Linear(16, 2)
    def forward(self, x):
        return self.fc2(self.act(self.fc(x)))

m_gelu = TinyMLP(use_gelu=True)
m_relu = TinyMLP(use_gelu=False)
print("Implementing paper components and comparing variants is the core of reproduction.")
```
