## 🧠 Concept: What are Transformer Pipelines?

**Pipelines** are high-level APIs that make using pre-trained models incredibly easy.

### Analogy: Using Pre-built Tools
Think of pipelines like using a smartphone:
- You don't need to understand the hardware
- You don't need to write the operating system
- You just use the apps!

Similarly, pipelines let you use powerful LLMs without:
- Training models from scratch
- Understanding all the technical details
- Writing complex code

### The Three Components
Every pipeline has three parts:

```
1. Tokenizer: Text → Numbers
2. Model: Numbers → Predictions
3. Post-processing: Predictions → Human-readable output
```

### Available Pipeline Types
- **sentiment-analysis:** Detect emotions in text
- **zero-shot-classification:** Classify without training
- **text-generation:** Generate new text
- **fill-mask:** Predict missing words
- **ner:** Named Entity Recognition
- **question-answering:** Answer questions from context
- **summarization:** Summarize long text
- **translation:** Translate between languages

---

````python
# Step 1: Install and Import Libraries
# Uncomment the line below if running in Colab or if transformers is not installed
# !pip install transformers torch

from transformers import pipeline
import warnings
warnings.filterwarnings('ignore')

print("✅ Transformers library imported successfully!")
print("📦 Ready to use pre-trained models!")
````

## 🎭 Part 1: Sentiment Analysis

**Sentiment Analysis** detects emotions in text (positive, negative, neutral).

### Use Cases:
- Customer feedback analysis
- Social media monitoring
- Product review analysis
- Brand reputation tracking

### Analogy: Emotional Intelligence
Like Lord Brahma understanding the emotions of all beings, sentiment analysis understands the emotional tone of text.

---

````python
# Step 2: Sentiment Analysis Pipeline

print("🎭 Creating Sentiment Analysis Pipeline...")
sentiment_classifier = pipeline("sentiment-analysis")
print("✅ Pipeline ready!\n")

# Test with different texts
texts = [
    "I am so happy with this new app Kiara - Money, Budget & Expense Tracker App at nexageapps.com",
    "This product is terrible and doesn't work at all.",
    "The weather is okay today, nothing special.",
    "Absolutely amazing! Best purchase ever!",
    "I'm disappointed with the customer service."
]

print("📊 Sentiment Analysis Results:\n")
print("="*70)

for text in texts:
    result = sentiment_classifier(text)[0]
    
    # Format output
    sentiment = result['label']
    confidence = result['score']
    
    # Add emoji based on sentiment
    emoji = "😊" if sentiment == "POSITIVE" else "😞"
    
    print(f"Text: {text[:60]}..." if len(text) > 60 else f"Text: {text}")
    print(f"Sentiment: {emoji} {sentiment} (Confidence: {confidence:.2%})")
    print("-"*70)
````

## 🏷️ Part 2: Zero-Shot Classification

**Zero-shot Classification** assigns labels to text without any training examples!

### How It Works:
You provide:
1. Text to classify
2. Possible labels

The model figures out which label fits best.

### Analogy: Universal Understanding
Like Brahma having broad vision to categorize all of creation, zero-shot models can categorize text into any labels you provide.

---

````python
# Step 3: Zero-Shot Classification Pipeline

print("🏷️ Creating Zero-Shot Classification Pipeline...")
zero_shot_classifier = pipeline("zero-shot-classification")
print("✅ Pipeline ready!\n")

# Example: Classify user feedback
text = "I am so happy with this new app Kiara - Money, Budget & Expense Tracker App at nexageapps.com"
labels = ["feature request", "user feedback", "bug report", "question", "complaint"]

result = zero_shot_classifier(text, labels)

print("📝 Text to classify:")
print(f"   {text}\n")
print("🎯 Classification Results:\n")
print("="*70)

for label, score in zip(result['labels'], result['scores']):
    bar = "█" * int(score * 50)
    print(f"{label:20s} {bar} {score:.2%}")

print("="*70)
print(f"\n✅ Best match: {result['labels'][0]} ({result['scores'][0]:.2%})")
````

## ✍️ Part 3: Text Generation

**Text Generation** creates new text based on a prompt.

### How It Works:
- You provide a starting prompt
- Model predicts next words
- Continues until max length or stop condition

### Parameters:
- **max_length:** Maximum tokens to generate
- **num_return_sequences:** Number of different outputs
- **temperature:** Randomness (higher = more creative)

---

````python
# Step 4: Text Generation Pipeline

print("✍️ Creating Text Generation Pipeline...")
generator = pipeline("text-generation", model="distilgpt2")
print("✅ Pipeline ready!\n")

# Generate text
prompt = "In nexageapps.com, we will teach you how to"

print(f"📝 Prompt: {prompt}\n")
print("🤖 Generated Texts:\n")
print("="*70)

results = generator(
    prompt,
    max_length=50,
    num_return_sequences=3,
    temperature=0.8
)

for i, result in enumerate(results, 1):
    print(f"\nGeneration {i}:")
    print(result['generated_text'])
    print("-"*70)
````

## 🎯 Part 4: Fill-in-the-Blank (Masked Language Modeling)

**Fill-Mask** predicts missing words in a sentence.

### How It Works:
- Replace a word with `<mask>`
- Model predicts what word should go there
- Returns top predictions with confidence scores

### Use Cases:
- Autocomplete
- Spell checking
- Text suggestions

---

````python
# Step 5: Fill-Mask Pipeline

print("🎯 Creating Fill-Mask Pipeline...")
fill_mask = pipeline("fill-mask")
print("✅ Pipeline ready!\n")

# Test with masked text
text = "This course will teach you all about <mask> models."

print(f"📝 Text with mask: {text}\n")
print("🔮 Top Predictions:\n")
print("="*70)

results = fill_mask(text, top_k=5)

for i, result in enumerate(results, 1):
    predicted_word = result['token_str'].strip()
    score = result['score']
    complete_text = result['sequence']
    
    print(f"{i}. Word: '{predicted_word}' (Confidence: {score:.2%})")
    print(f"   Complete: {complete_text}")
    print("-"*70)
````

## 👤 Part 5: Named Entity Recognition (NER)

**NER** identifies and classifies named entities in text.

### Entity Types:
- **PER:** Person names
- **ORG:** Organizations
- **LOC:** Locations
- **MISC:** Miscellaneous

### Use Cases:
- Information extraction
- Resume parsing
- Content categorization

---

````python
# Step 6: Named Entity Recognition Pipeline

print("👤 Creating NER Pipeline...")
ner = pipeline("ner", grouped_entities=True)
print("✅ Pipeline ready!\n")

# Test with text containing entities
text = "My name is Karthik Arjun. I am a Master's in AI student at University of Auckland. I work at Deloitte, NZ!"

print(f"📝 Text: {text}\n")
print("🔍 Detected Entities:\n")
print("="*70)

results = ner(text)

for entity in results:
    entity_text = entity['word']
    entity_type = entity['entity_group']
    confidence = entity['score']
    
    # Add emoji based on entity type
    emoji_map = {
        'PER': '👤',
        'ORG': '🏢',
        'LOC': '📍',
        'MISC': '🏷️'
    }
    emoji = emoji_map.get(entity_type, '❓')
    
    print(f"{emoji} {entity_text:25s} → {entity_type:10s} (Confidence: {confidence:.2%})")

print("="*70)
````

## ❓ Part 6: Question Answering

**Question Answering** extracts answers from a given context.

### How It Works:
1. Provide context (paragraph/document)
2. Ask a question
3. Model finds the answer in the context

### Use Cases:
- Document search
- Customer support bots
- Educational tools

---

````python
# Step 7: Question Answering Pipeline

print("❓ Creating Question Answering Pipeline...")
qa = pipeline("question-answering")
print("✅ Pipeline ready!\n")

# Context and questions
context = """My name is Karthik Arjun. I am a Master's in AI student at University of Auckland. 
I work at Deloitte, NZ! He has 12.5 years of SAP experience. He is a certified SAP BTP & Integration Architect!"""

questions = [
    "Where do I work?",
    "What is my name?",
    "How many years of experience do I have?",
    "What am I studying?",
    "What certifications do I have?"
]

print("📄 Context:")
print(context)
print("\n" + "="*70)
print("\n❓ Questions & Answers:\n")

for question in questions:
    result = qa(question=question, context=context)
    
    answer = result['answer']
    confidence = result['score']
    
    print(f"Q: {question}")
    print(f"A: {answer} (Confidence: {confidence:.2%})")
    print("-"*70)
````

## 📝 Part 7: Text Summarization

**Summarization** condenses long text into shorter versions while preserving key information.

### Types:
- **Extractive:** Selects important sentences
- **Abstractive:** Generates new summary text

### Use Cases:
- News article summaries
- Document digests
- Meeting notes

---

````python
# Step 8: Summarization Pipeline

print("📝 Creating Summarization Pipeline...")
summarizer = pipeline("summarization")
print("✅ Pipeline ready!\n")

# Long text to summarize
long_text = """My name is Karthik Arjun. I am a Master's in AI student at University of Auckland. 
I work at Deloitte, NZ! He has 12.5 years of SAP experience. He is a certified SAP BTP & Integration Architect! 
Throughout his career, he has worked on numerous enterprise integration projects, helping organizations 
streamline their business processes and adopt cloud technologies. His expertise spans across SAP Business 
Technology Platform, integration patterns, and AI/ML applications in enterprise systems."""

print("📄 Original Text:")
print(long_text)
print(f"\nLength: {len(long_text)} characters\n")
print("="*70)

# Generate summary
summary = summarizer(long_text, max_length=60, min_length=20, do_sample=False)

print("\n📋 Summary:")
print(summary[0]['summary_text'])
print(f"\nLength: {len(summary[0]['summary_text'])} characters")
print(f"Compression: {len(summary[0]['summary_text']) / len(long_text):.1%}")
````

## 🌍 Part 8: Translation

**Translation** converts text from one language to another.

### Available Models:
- Helsinki-NLP models (100+ language pairs)
- Facebook's M2M models (multilingual)
- Google's T5 models

---

````python
# Step 9: Translation Pipeline

print("🌍 Creating Translation Pipeline (French → English)...")
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
print("✅ Pipeline ready!\n")

# French text to translate
french_text = "Ce cours est produit par Karthik. Refer from HF portal"

print(f"🇫🇷 French: {french_text}\n")

# Translate
result = translator(french_text)

print(f"🇬🇧 English: {result[0]['translation_text']}")
````
