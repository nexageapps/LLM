## 🧠 Concept: The Position Problem

**Problem:** Transformers process all tokens simultaneously (parallel processing). They don't know the order of words!

### Example:
```
"Dog bites man" vs "Man bites dog"
```
Without position info, these look the same to a transformer!

### Solution: Positional Encoding
Add position information to each token's embedding:

```
Final Embedding = Word Embedding + Positional Encoding
```

### Sinusoidal Encoding Formula

$$PE_{(pos, 2i)} = \sin\left(\frac{pos}{10000^{2i/d}}\right)$$

$$PE_{(pos, 2i+1)} = \cos\left(\frac{pos}{10000^{2i/d}}\right)$$

Where:
- $pos$: Position in sequence
- $i$: Dimension index
- $d$: Embedding dimension

---

````python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn

print("✅ Libraries imported!")
````

````python
# Step 2: Implement Sinusoidal Positional Encoding

def get_positional_encoding(seq_len, d_model):
    """
    Generate sinusoidal positional encodings.
    
    Args:
        seq_len: Sequence length
        d_model: Embedding dimension
    
    Returns:
        Positional encoding matrix (seq_len, d_model)
    """
    position = np.arange(seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    
    pe = np.zeros((seq_len, d_model))
    pe[:, 0::2] = np.sin(position * div_term)
    pe[:, 1::2] = np.cos(position * div_term)
    
    return pe

# Generate positional encodings
seq_len = 100
d_model = 512

pe = get_positional_encoding(seq_len, d_model)

print(f"✅ Positional encoding shape: {pe.shape}")
print(f"Sequence length: {seq_len}")
print(f"Embedding dimension: {d_model}")
````

````python
# Step 3: Visualize Positional Encodings

plt.figure(figsize=(15, 8))

# Plot heatmap
plt.subplot(2, 1, 1)
plt.imshow(pe, cmap='RdBu', aspect='auto')
plt.colorbar(label='Encoding Value')
plt.xlabel('Embedding Dimension')
plt.ylabel('Position in Sequence')
plt.title('Sinusoidal Positional Encoding Heatmap')

# Plot specific positions
plt.subplot(2, 1, 2)
positions_to_plot = [0, 10, 20, 50]
for pos in positions_to_plot:
    plt.plot(pe[pos, :50], label=f'Position {pos}')
plt.xlabel('Dimension')
plt.ylabel('Encoding Value')
plt.title('Positional Encoding Patterns (First 50 dimensions)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📊 Visualization complete!")
````

## 🎓 Key Takeaways

1. Transformers need positional encoding to understand word order
2. Sinusoidal encoding uses sine and cosine functions
3. Each position gets a unique encoding pattern
4. Positional encodings are added to word embeddings
5. Alternative: Learned positional embeddings (BERT)
